## 1. Area Plot

**What is a line plot and why use it?**

An Area chart or area plot is a type of plot which displays information as a series of data points called 'markers' connected by straight line segments. It is a basic type of chart common in many fields. Use line plot when you have a continuous data set. These are best suited for trend-based visualizations of data over a period of time.

**Questions:**
`Berapa pergerakan jumlah kasus kriminalitas berupa perampokan tahun 2008 - 2016?`

```python
# Akan dibuat visualisasi mengenai jumlah kriminalitas berupa perampokan tahun 2008 - 2016

# pindahkan kolom major_category menjadi index
df_lndcrime = df_lc.set_index("major_category")
df_lndcrime
```

```python
df_bul = df_lndcrime.loc["Burglary"]
df_bul
```

```python
df_bul = df_bul.drop(columns=['lsoa_code', 'borough', 'minor_category', 'month'])
df_bul
```

```python
df_bul = df_bul.set_index('year')
df_bul
```

```python
df_area = df_bul.groupby('year').sum()
df_area = df_area.rename(columns={"value":"Burglary"})
df_area
```

```python
df_area.plot(kind='area', figsize=(15,7))

plt.title("Kasus Perampokan di London")
plt.xlabel("Tahun")
plt.ylabel("Value")
```

**Questions:**
`Bagaimana perbandingan pergerakan jumlah kriminalitas berupa pencurian dan obat-obatan terlarang tahun 2008 - 2016?`

```python
# Akan dibuat visualisasi mengenai jumlah kriminalitas berupa pencurian dan obat-obatan terlarang tahun 2008 - 2016
df_cd = df_lndcrime.loc["Criminal Damage"]
df_cd = df_cd.drop(columns=['lsoa_code', 'borough', 'minor_category', 'month'])
df_cd = df_cd.set_index('year')
df_cd = df_cd.groupby('year').sum()
df_cd = df_cd.rename(columns={"value":"Criminal Damage"})
df_cd
```

```python
# gabungkan
df_area["Criminal Damage"] = df_cd["Criminal Damage"]

df_area
```

```python
df_area.plot(kind='area', figsize=(13,7), stacked=False)

plt.title("Burgarly and Criminal Damage Case in London")
plt.xlabel("Tahun")
plt.ylabel("Value")
```

## 2. Bar Chart

A bar plot is a way of representing data where the *length* of the bars represents the magnitude/size of the feature/variable. Bar graphs usually represent numerical and categorical variables grouped in intervals. 

To create a bar plot, we can pass one of two arguments via `kind` parameter in `plot()`:

* `kind=bar` creates a *vertical* bar plot
* `kind=barh` creates a *horizontal* bar plot

**Question:**

`Bagaimana jumlah kriminalitas pada tahun 2014 untuk setiap macam kejahatan di London?`

```python
# Akan dibuat visualisasi mengenai jumlah kriminalitas pada tahun 2008 - 2016 
# untuk setiap major_category pada tahun 2015

df_2014 = df_lndcrime[df_lndcrime["year"]==2014]
df_2014
```

```python
df_bar = df_2014.groupby('major_category').sum().drop(columns=['year','month']).rename(columns={"value":"2014"})
df_bar
```

```python
df_bar.plot(figsize=(13,5), kind="bar")

plt.title("2014 Criminal Case in London")
plt.xlabel("Major Category")
plt.ylabel("Value")
```

**Question:**

`Bagaimana jumlah kriminalitas pada tahun 2014 - 2016 untuk setiap macam kejahatan di London?`

```python
# Akan dibuat visualisasi mengenai jumlah kriminalitas 2014, 2015, dan 2016

df_2015 = df_lndcrime[df_lndcrime["year"]==2015]
df_2015 = df_2015.groupby('major_category').sum().drop(columns=['year','month']).rename(columns={"value":"2015"})
df_2015
```

```python
df_2016 = df_lndcrime[df_lndcrime["year"]==2015]
df_2016 = df_2016.groupby('major_category').sum().drop(columns=['year','month']).rename(columns={"value":"2016"})
df_2016
```

```python
df_bar.plot(kind='bar', figsize=(13,7))

plt.title("2014-2016 Criminal Case in London")
plt.xlabel("Major Category")
plt.ylabel("Value")
```

## 5. Box Plot

A `box plot` is a way of statistically representing the *distribution* of the data through five main dimensions: 

- **Minimun:** Smallest number in the dataset.
- **First quartile:** Middle number between the `minimum` and the `median`.
- **Second quartile (Median):** Middle number of the (sorted) dataset.
- **Third quartile:** Middle number between `median` and `maximum`.
- **Maximum:** Highest number in the dataset.
